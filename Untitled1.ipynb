##### ייבוא ספריות רלוונטיות

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
url='https://raw.githubusercontent.com/giliandtal/project/main/executions.csv'
ex_df=pd.read_csv(url)
ex_df
```

```python
ex_df.shape
```

```python
ex_df.columns
```

```python
ex_df.isnull().sum()
```

```python
ex_df.dtypes
```

```python
numeric_features=['Age','Victim Count']
categorical_features=['Sex','Race','Foreign National']
target_features=['Method','Region','Victim Sex']
all_fe=numeric_features+categorical_features+target_features
all_fe
```

```python
ex_df_dum = pd.get_dummies(ex_df[all_fe])
ex_df_dum.columns
```

```python
correlation_matrix = ex_df_dum.corr(numeric_only = False, method = 'spearman' ).round(2)
correlation_matrix
```

```python
plt.figure(figsize=(17,17))
sns.heatmap(data=correlation_matrix,cmap='coolwarm', annot=True, xticklabels=ex_df_dum.columns.tolist(), yticklabels=target_features)
plt.show()
```
