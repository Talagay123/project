```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
url='https://raw.githubusercontent.com/giliandtal/project/main/pakistan_cars_1990-2006.csv'
url2='https://raw.githubusercontent.com/giliandtal/project/main/pakistan_cars_2007-2012.csv'
url3='https://raw.githubusercontent.com/giliandtal/project/main/pakistan_cars_2013-2016.csv'
url4='https://raw.githubusercontent.com/giliandtal/project/main/pakistan_cars_2017-2019.csv'
url5='https://raw.githubusercontent.com/giliandtal/project/main/pakistan_cars_2020-2024.csv'
```

```python
pak1_df=pd.read_csv(url)
pak2_df=pd.read_csv(url2)
pak3_df=pd.read_csv(url3)
pak4_df=pd.read_csv(url4)
pak5_df=pd.read_csv(url5)
pak_df=pd.concat([pak1_df,pak2_df,pak3_df,pak4_df,pak5_df]).reset_index()
```

```python
ax=sns.countplot(data=pak_df, x="model")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=6)
plt.show
```

```python
plt.figure(figsize=(10, 6))
ax=sns.stripplot(x='model', y='price', data=pak_df,jitter=0.3)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=7)
plt.show
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=pak_df, x='assembly',palette='Set2')
plt.title('Count of Cars by Assembly Type')
plt.xlabel('Assembly Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax=sns.boxplot(x='assembly', y='price', data=pak_df, palette='Set2')
ax.set_yscale("log")
```

```python
pak_df['manufacturer'] = pak_df['title'].str.split(' ').str[0]
manufacturer_counts = pak_df['manufacturer'].value_counts().head(15)
plt.figure(figsize=(12, 8))
sns.barplot(x=manufacturer_counts.values, y=manufacturer_counts.index, palette="viridis")
```

```python
top_vehicles = pak_df['title'].value_counts().head(15)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_vehicles.values, y=top_vehicles.index, palette="viridis")

```

ניתן לראות מהגרף הראשון שיצרנית הרכב המובילה בפקיסטן היא טויוטה והבאות אחריה הן סוזוקי והונדה. גם רוב החברות שמופיעות לאחר מכן - רובן חברות יםניות.
ניתן להסיק שרכבים יפנים הם המועדפים על התושבים בפקיסטן ובפרט טויוטה סוזוקי והונדה.

אמנם טויוטה היא היצרנית המובילה אבל דווקא דגמים של הונדה סיוויק הם הנסחרים ביותר

```python
top_vehicles = pak_df['title'].value_counts().head(15).index

top_vehicles_df = pak_df[pak_df['title'].isin(top_vehicles)]
avg_mileage = top_vehicles_df.groupby('title')['mileage'].mean()
vehicle_counts = top_vehicles_df['title'].value_counts()
result_df = pd.DataFrame({'average_mileage': avg_mileage, 'count': vehicle_counts})
sorted_result_df = result_df.sort_values(by='count', ascending=False)
print(sorted_result_df)
```

```python
top_vehicles = pak_df['title'].value_counts().head(15).index

top_vehicles_df = pak_df[pak_df['title'].isin(top_vehicles)]
avg_price = top_vehicles_df.groupby('title')['price'].mean()
vehicle_counts = top_vehicles_df['title'].value_counts()
result_df = pd.DataFrame({'average_price': avg_price, 'count': vehicle_counts})
sorted_result_df = result_df.sort_values(by='count', ascending=False)
sorted_result_df['average_price'] = sorted_result_df['average_price'].round()

print(sorted_result_df)
```

```python
plt.figure(figsize=(10, 6))
ax=sns.boxplot(y='mileage', data=pak_df)

ax.set_yscale("log")
```

```python
crosstab = pd.crosstab(pak_df['price_category'], pak_df['mileage_category'])
plt.figure(figsize=(10, 6))
sns.heatmap(crosstab, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Heatmap of Price Category vs. Mileage Category')
plt.xlabel('Mileage Category')
plt.ylabel('Price Category')

plt.show()
```

```python
filtered_df = pak_df[(pak_df['fuel_type'] != 'Not Available')]
plt.figure(figsize=(10, 6))
ax=sns.boxplot(x='fuel_type', y='mileage', data=filtered_df,palette='viridis')

plt.title('Box Plot of Mileage by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Mileage')

plt.show()
```

סוג הדלק הזול ביותר בפקיססטן הוא "High Speed Diesel"
בגרף זה ניתן לראות שהתפלגות מספר הקילומטרים בדיזל היא יותר רחבה.


```python
filtered_df = pak_df[(pak_df['price'] != 0) & (pak_df['fuel_type'] != 'Not Available')]
plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_df, x='fuel_type', y='price',palette='viridis')
plt.yscale('log')
plt.title('Relationship between Fuel Type and Price')
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.show()
```

 מבחינת מחיר- מחירי הרכבים שיש להם דלק דיזל- יותר יקרים.

```python
city_counts = pak_df['city'].value_counts().head(15)
plt.figure(figsize=(12, 8))
sns.barplot(x=city_counts.index, y=city_counts.values, palette='viridis')

plt.title('Top 15 Cities with the Most Cars')
plt.xlabel('City')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()
           
```

```python
top_cities = pak_df.groupby('city')['price'].count().nlargest(3).index
plt.figure(figsize=(10,6))
sns.boxplot(data=pak_df[pak_df['city'].isin(top_cities)], x='city', y='price')
plt.title('Boxplot of Car Prices in Top 3 Cities with Highest Car Counts')
plt.yscale('log')
plt.xlabel('City')
plt.ylabel('Price')
plt.show()
```
